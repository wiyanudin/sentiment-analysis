```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.telkom.tracencare',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=100000,
    filter_score_with=None
)
```

```python
df_pl = pd.DataFrame(np.array(result),columns=['review'])

df_pl = df_pl.join(pd.DataFrame(df_pl.pop('review').tolist()))

df_pl.head()
```

```python
len(df_pl.index)
```

```python
df_pl[['userName', 'score','at', 'content']].head()
```

```python
new_df = df_pl[['userName', 'score','at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=False)
sorted_df.head()
```

```python
my_df = sorted_df[['userName', 'score','at', 'content']]
```
